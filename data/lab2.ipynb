## 2.2 studentId

- length of 9 digits
- type: int

````python
# Define function to generate correct id
def generate_student_id(n=8):
    # Generate a random number whose first leading number is not 0（1-9）
    first_digit = random.choice('123456789')
    # Generate a random last 8 digits
    random_digits = ''.join(random.choices('0123456789', k=n))
    # Combined into a complete student ID
    student_id = first_digit + random_digits
    return student_id
````

# 4 Validate exist errorse

Inspection conditions:
1. studentId - length of 9 digits
2. courseId - legnth of 7 characters, 3 letters followed by 4 digits
3. term - one of "WINTER", "SUMMER", 'FALL'
4. year - 1965 -2025

Possible errors:
1. studentId, Length is not 9 digits
2. courseId, should be 3 letters followed by 4 digits
3. term, beyond WINTER, SUMMER and FALL
4. year, less than 1965 or big than 2025

In fact, the second and third lines in the examples already contain the errors described above.

````python
# Validation function
def validate_data(df):
    errors = []

    for idx, row in df.iterrows():
        student_id = row['studentId']
        course_id = row['courseId']
        term = row['term']
        year = int(row['year'])

        if len(student_id) != 9:
            errors.append((idx, 'studentId', student_id, 'Length is not 9 digits'))
        if not re.match(r'^[A-Za-z]{3}\d{4}$', course_id):
            errors.append((idx, 'courseId', course_id, 'Invalid format (should be 3 letters followed by 4 digits)'))
        if term not in termList:
            errors.append((idx, 'term', term, 'Invalid term (should be one of WINTER, SUMMER, FALL)'))
        if year < 1965 or year > 2025:
            errors.append((idx, 'year', year, 'Invalid year (should be between 1965 and 2025)'))

    return errors
````

````python
# Validate data
validation_errors = validate_data(final_df)

# Print validation errors
for error in validation_errors:
    print(f'Row {error[0]}: Column {error[1]} - Value "{error[2]}" - Error: {error[3]}')
````

# 5 Create new errors

````python
# Validate existing errors
initial_validation_errors = validate_data(final_df)

# Find rows number of exist errors
error_rows_existing = {error[0] for error in initial_validation_errors}

# Calulate rows for new errors
num_errors_to_generate = 10 - len(error_rows_existing)

# Create errors in rest rows without errors
remaining_indices = set(range(1, len(final_df))) - error_rows_existing
error_rows_to_generate = random.sample(sorted(remaining_indices), num_errors_to_generate)

for idx in error_rows_to_generate:
    # Randomly choose 1 to 4 error types
    num_errors = random.randint(1, 2)
    error_types = random.sample(['studentId', 'courseId', 'term', 'year'], num_errors)
    print(idx, error_types)
    if 'studentId' in error_types:
        student_id = final_df.at[idx, 'studentId']
        if random.choice([True, False]):
            # Remove 1 or 2 digits
            num_to_remove = random.randint(1, 2)
            final_df.at[idx, 'studentId'] = student_id[:-num_to_remove]
        else:
            # Add 1 random digit
            # num_to_add = random.randint(1, 2)
            additional_digits = ''.join(random.choices('0123456789', k=1))
            final_df.at[idx, 'studentId'] = student_id + additional_digits
        print(f'studentId modified to {final_df.at[idx, "studentId"]}')
        
    if 'courseId' in error_types:
        final_df.at[idx, 'courseId'] = final_df.at[idx, 'courseId'] + 'X'  # Make courseId invalid
        print('courseId')
        
    if 'term' in error_types:
        final_df.at[idx, 'term'] = random.choice(['AUTUMN', 'SPRING'])  # Make term invalid
        print('term')        
        
    if 'year' in error_types:
        final_df.at[idx, 'year'] = random.choice([1956, 1957, 1958, 1959, 1960, 2030, 2031, 2032, 2033])  # Make year out of range
        print('year')      

# Validate data
validation_errors = validate_data(final_df)

# Print validation errors
for error in validation_errors:
    print(f'Row {error[0]}: Column {error[1]} - Value "{error[2]}" - Error: {error[3]}')
````
